## Section 1 - importing packages

```python
# package imports
import pygame
```

## Section 2 - setting up the game window and other assets

```python
# setting up the game window
width, height = 400, 300
screen = pygame.display.set_mode((width, height))
pygame.display.set_caption('My First Pygame Window')

# adding images for sprites
bird_image = pygame.image.load('bird.png')  # Make sure to have an image named 'bird.png' in the same folder
bird_image = pygame.transform.scale(bird_image, (50, 50))
```

## Section 3 - running the main game loop

```python
# Main game loop
running = True

while running:
    # Event handling
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False

    # *** THIS IS WHERE WE DRAW THINGS ON THE SCREEN & UPDATE THE GAME ENVIRONMENT ***
    
    # ***
    
    # The .flip() function will update the screen with all the changes we made in the starred section above
    # changes are not visible to the user until we call .flip()
    pygame.display.flip()
    
    # This is the framerate in which we update our game!
    pygame.time.delay(30)
    
# Quit Pygame
pygame.quit()
```
